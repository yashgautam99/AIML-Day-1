```python
def takingInput():
    A=int(input("Enter First number:"))
    B=int(input("Enter Second number:"))
    op=input("Enter operator(+,-,*,/):")
    f=0
    while(f==0):
      if(op=='+' or op=='-' or op=='*' or op=='/'):
        f=1
        print('Right operator selected')
      else:
        f=0
        print('Wrong operaton selcted')
    return([A,B,op])
```
